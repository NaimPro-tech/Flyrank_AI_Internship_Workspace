## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Reasoning
- Our dataset contains many interacting search signals such as impressions, clicks, CTR, position, and search volume. A fixed if-else rule cannot capture non linear relationships between these features and may fail when new data patterns appear. Machine learning learns these patterns from data and assigns a risk score to every page, making the recommendations more reliable and scalable.

- Since the decision depends on multiple search signals rather than a single condition, a fixed if-statement is not sufficient.
