```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between Search Volume and 90 days Impressions; {corr:.3f}")
print("Observation\n" \
"In this dataset, we observed almost no linear relationship between these two variables. ")
print("So lets find out group wise relation.")
position_search = df.groupby("position_tier").agg(avg_search_volume=("search_volume","mean"),
                                                  avg_impression=("impressions_90d", "mean"),
                                                  avg_clicks=("clicks_90d", "mean"))

print("Position wise analysis:\n ", position_search.sort_values(by="position_tier", ascending=False))
print("Observation of Position wise analysis:\n" \
"We observed that pages in the page_3_5 and striking tiers receive lower average clicks than page_1,\n" \
"despite being associated with higher search-volume queries.")
```

```python
analysis_ = df.groupby("position_tier").agg(avg_search_volume=("search_volume", "mean"),
                                            total_ctr=("ctr", "mean"))
print(analysis_.sort_values(by="position_tier",ascending=False))
print("Observation\nThe ctr varies with positions but top 3 pages have lower average search volume but higher CTR.\n" \
"This raises an important question: What earns their ctr?")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

## What I can claim
- Observed Trends & Patterns: I can show from the data which position tiers or categories are seeing the biggest drops in clicks and impressions.
- Directional Insights: Model can guidance on which content updates have the highest probability of regaining traffic.
- Decision-Support Tool: It can be decision-support tool to SEO team to make a content refresh priority list with data driven insights.

## What I can't claim
- Casual proof: Model can't 100% sure on any specific observation.Data can only show correlation, not causation.
- Predicting Google: I don't predict the hidden search algorithm of google, also the work of my model isn't to predict how google change the rank of a content.
- Guaranteed Traffic Recovery: Model can't guarantee that only refresh content can regain traffic. It can only increase the probability of success.
